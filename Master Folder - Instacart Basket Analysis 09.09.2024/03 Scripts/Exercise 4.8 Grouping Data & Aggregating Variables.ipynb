# 2. Grouping Data & Aggregating Variables

### 2.1 Group the data and find the aggregated mean

```python
#Group the data frame by the department_id column and aggregate the mean of the order_number column

df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

### How do the results for the entire data frame differ from those of the subset?
In my case, the subset output had already shown all the department numbers, not just a part, since my data frame seems to be ordered diffrently. However, taking the output from the exercise brief into account for the comparison shows that the mean for 4(produce) slightly dropped to an average of 18; on the other hand, the mean for 17 (household) increased to 16, which means that produce does not sell much more than products of the household department. Products of department 21(missing) sell the most on an average of 23, and products belonging to the alcohol department (5) sell the least by 15, whereas other departments are close to selling the same.

### 2.2 Creating a loyalty flag for existing customers

```python
#Creating a loyalty flag for existing customers by using the transform() and loc () functions

#Start with the transform function by applying the max to the order_number column, and creating a new column called "max_order".

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
#Next, executing the loc() function to set certain conditions for the loyalty_flag column

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts()
```

### 2.3 Spending habits of the three types of customers

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['max', 'mean', 'min']})
```

### 2.4 Creating a spending flag for customers

```python
#Creating a spending flag for existing customers by using the transform() and loc () functions

#Start with the transform function by applying the mean to the prices column, and creating a new column called "average_prices".

df_ords_prods_merge['average_prices'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
#Next, executing the loc() function to set certain conditions for the spending_flag column

df_ords_prods_merge.loc[df_ords_prods_merge['average_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge['spending_flag'].value_counts()
```

### 2.5 Creating a frequncy flag for days since last order column

```python
#Creating a frequency flag for existing customers by using the transform() and loc () functions

#Start with the transform function by applying the median to the days since last order column, and creating a new column called "median_days_since_last_order".

df_ords_prods_merge['median_days_since_last_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform('median')
```

```python
#Next, executing the loc() function to set certain conditions for the frequency_flag column

df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_last_order'] >= 10 , 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_last_order'] >10) & (df_ords_prods_merge['median_days_since_last_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_last_order'] > 20 , 'frequency_flag'] = 'Non-frequent customer'
```

```python
#Treating the outliers of the prices column, any value above 100 is replaced with NaN value.

df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
