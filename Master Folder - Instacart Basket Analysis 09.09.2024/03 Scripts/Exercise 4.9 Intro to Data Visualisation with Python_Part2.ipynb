### 2.5 Create a histogram of the "order_hour_of_day" column

```python
# Create a histogram

df_ords_prods_all['order_hour_of_day'].plot.hist(bins = 50)
```

__Comment:__
The histogram demonstrates that most orders are placed in the morning around 10:00 am and in the afternoon from 12:00 pm to 16:00 pm. There are no orders placed somewhere around 11:00 am - 12:00 pm and 5:00 pm and 6:00 pm. However, the bins are not accurately assigned to the bar charts which makes it difficult to describe the histogram to its fullest.

```python
df_ords_prods_all['order_hour_of_day'].value_counts()
```

```python
# Adjust the histogram for more accuracy
# Define the bin edges from 0 to 24

bins = np.arange(0,25)
```

```python
# Show the histogram for the correct adjustment

df_ords_prods_all['order_hour_of_day'].plot.hist(bins=bins, edgecolor='black')
```

```python
# Add labels and titles
fig2, ax2 = plt.subplots()
df_ords_prods_all['order_hour_of_day'].plot.hist(bins=bins, edgecolor='black')
ax2.set_xticks(np.arange(0, 24, 2))
ax2.set_xlabel('Hour of the Day', fontsize=15)
ax2.set_ylabel('Frequency', fontsize=15)
ax2.set_title('Distribution of Orders by Hour of Day', fontsize=18)
```

__Comment:__ Looking at the adjusted histogram shows that the most orders are placed around 11:00 am. In general the hours from 09:00 am until 06:00 pm is the busiest period of the day. Orders are placed around-the-clock.

```python
# Exporting the histogram to an image file for further use in presentations

fig2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'hist_ord_hour_day.png'))
```

### 2.9 Create a scatterplot to show the connection between age and spending power

```python
# Create a subset for the entire data frame, including the necessary columns for creating the line chart

df_ords_prods_all_subset_3 = df_ords_prods_all[['age', 'income']]
```

```python
# Create a scatterplot

fig6, ax6 = plt.subplots()
sns.scatterplot(x = 'age', y = 'income', data = df_ords_prods_all_subset_3)
ax6.set_xlabel('Age', fontsize=15)
ax6.set_ylabel('Income', fontsize=15)
ax6.set_title('Connection between Age and Spending Power', fontsize=18)
```

```python
# Exporting the scatterplot to an image file for further use in presentations

fig6.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'connection between age and spending power.png'), bbox_inches='tight')
```

__Comment:__ Connection between Age and Spending Power

The scatter plot above visualizes the relationship between customers' ages and their income (spending power). The key observations from the plot are:

- There appears to be a positive correlation between age and income. As age increases, income tends to rise, particularly through the middle age range.
- Younger individuals typically show lower income levels, while older individuals, particularly those in their 30s to 50s, tend to earn more.
- The income level seems to stabilize or even slightly decline in the older age groups (possibly due to retirement or reduced working hours).
