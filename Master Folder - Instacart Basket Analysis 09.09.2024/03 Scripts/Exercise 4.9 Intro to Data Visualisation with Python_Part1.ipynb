## 3. Merging the data

```python
# Import the data for merging

df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_new.pkl'))
```

```python
# Check data frame for correct import

df_ords_prods_merged
```

```python
# Check for identical data type for the mergeing process on the user_id column

df_ords_prods_merged.info()
```

```python
# Reference for correct merge

df_customers.shape
```

```python
# Merge the data sets df_ords_prods_merged and df_customers on the user_id column with an inner join

df_ords_prods_all = df_ords_prods_merged.merge(df_customers, on ='user_id', indicator = 'merge_indicator')
```
