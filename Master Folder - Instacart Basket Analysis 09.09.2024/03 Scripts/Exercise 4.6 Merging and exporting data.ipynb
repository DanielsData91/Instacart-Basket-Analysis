## 4.6 Merging and exporting data

### This script contains the following points:

#### 1. Import the data
#### 2. Merge data
#### 3. Exporting the data

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

#### 1. Import the data

```python
path = r'/Users/daniel/Desktop/LEBEN/Data Analyst/Data Analytics Course/Data Immersion/Achievement 4/Master Folder - Instacart Basket Analysis 09.09.2024'
```

```python
#Importing the products.csv data from the original data folder
df_ords_prior = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'orders_products_prior.csv'), index_col = False)
```

```python
#Check if the import was executed correctly
df_ords_prior
```

```python
#Importing the orders_cleaned data from the prepared data folder
df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_cleaned.csv'), index_col = 0)
```

```python
#Check if the import was executed correctly
df_ords
```

```python
#Drop the column 'unnamed 0' which is unnecessary for further analysis.
df_ords = df_ords.drop(columns = ['Unnamed: 0'])
```

```python
df_ords
```

#### 2. Merge data

```python
#Check the shape of the data frame for later merging operations
df_ords_prior.shape
```

```python
#Check the shape of the data frame for later merging operations
df_ords.shape
```

```python
#Merge the two data sets df_ords_prior and df_ords to df_merged_large by using an inner join.
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', how = 'inner', indicator = True)
```

```python
#Check the merged data frame
df_merged_large.head()
```

```python
#Check the shape of the data frame to confirm the number of rows and columns after merging
df_merged_large.shape
```

```python
#Confirming the merge by using the flag '_merge'
df_merged_large['_merge'].value_counts()
```

##### Comment: Due to using an inner join, the output reflects that the key specified exist in both dataframes

#### 3. Exporting the data

```python
#Exporting the merged data frame to a pkl format
df_merged_large.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```
