# 2. If-Statments

### 2.1 If-Statements with User-Defined Functions - SUBSET

```python
#Creating a subset of the data frame to avoid any potential memory issues

df_subset = df_ords_prods_merge[:1000000]
```

```python
#Defining a function to add price labels to the products

def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
#Apply the function for the price labels.

df_subset['price_range'] = df_subset.apply(price_label, axis=1)
```

```python
#Checking the frequency of the new column

df_subset['price_range'].value_counts(dropna = False)
```

```python
df_subset['prices'].max()
```

### 2.2 If-Statements with the loc() Function - SUBSET

```python
#loc function to create a condition for the high-range products, above 15

df_subset.loc[df_subset['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
#loc function to create a condition for the mid-range products, between 5 - 15, and equal to 15

df_subset.loc[(df_subset['prices'] <= 15) & (df_subset['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
#loc function to create a condition for the low-range products, below and equal to 5

df_subset.loc[df_subset['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
#Check the frequency of the new column

df_subset['price_range_loc'].value_counts(dropna = False)
```

### 2.4 If-Statements with For-Loops - DATAFRAME

#### 2.4.1 Busiest days of the week

```python
#Check for the frequency of each day to determine which are the busiest and less busy days of the week

df_ords_prods_merge['orders_day_of_the_week'].value_counts(dropna = False)
```

```python
#Create a For-Loop function to create results to create a new column that displays the degree of capacity for each day of the week

result = []

for value in df_ords_prods_merge["orders_day_of_the_week"]:
    if value == 0:
        result.append("Bussiest day")
    elif value == 4:
        result.append("least busy")
    else:
        result.append("Regularly busy")
```

```python
#Check on the result of the for-loop function

result
```

```python
#Create a new column named busiest_day to display the degree of capacity per day

df_ords_prods_merge['busiest_day'] = result
```

```python
#Check for the frequency of the three labels

df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
#Busiest days update: Create a For-Loop function to create results to create a new column that displays the degree of capacity for each day of the week


result = []

for value in df_ords_prods_merge["orders_day_of_the_week"]:
    if value in [0, 1]:
        result.append("Busiest days")
    elif value in [3, 4]:
        result.append("Slowest days")
    else:
        result.append("Regular days")
```

```python
#Create a new column named busiest_day_update to display the degree of capacity per day

df_ords_prods_merge['busiest_day_update'] = result
```

```python
#Check the values for accuracy

df_ords_prods_merge['busiest_day_update']
```

```python
df_ords_prods_merge['busiest_day_update'].value_counts(dropna = False)
```

#### Observation, difference between before and after the update:
Taking two days for "busiest" and "slowest" and three days for the "regular" label into account creates a more accurate distribution of the capacity Instacart has on certain days. In addition, it makes a clearer understanding of when it is recommended to place more ads to engage with potential customers and on the other hand when ads are more likely to be overlooked due to less traffic.

#### 2.4.2 Busiest hours of the days

```python
#Check for the frequency of each hour to determine which are the busiest and less busy hours of the day

df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
#Create a For-Loop function to create results to create a new column that displays the degree of capacity for each hour of the day

result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result.append("Average orders")
    else:
        result.append("Fewest orders")
```

```python
result
```

```python
#Create a new column named busiest_hours_of_the_day to display the degree of capacity per hour

df_ords_prods_merge['busiest_hours_of_the_day'] = result
```

```python
df_ords_prods_merge['busiest_hours_of_the_day'].value_counts(dropna = False)
```
